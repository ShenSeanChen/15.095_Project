```python
def round_of_rating(number):
    """Round a number to the closest half integer.
    >>> round_of_rating(1.3)
    1.5
    >>> round_of_rating(2.6)
    2.5
    >>> round_of_rating(3.0)
    3.0
    >>> round_of_rating(4.1)
    4.0"""

    return round(number * 2) / 2
```

## Round number of hours worked yesterday to nearest 0.5

```python
for index, row in data.iterrows():
    time = round_of_rating(row['Answer.work.yesterday.time'])
    data.at[index, 'work_yesterday_cleaned'] = time
```
